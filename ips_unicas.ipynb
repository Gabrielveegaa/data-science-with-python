```python
import pandas as pd

# Cargar los dos CSVs
df_logs = pd.read_csv("log.csv")
df_maliciosas = pd.read_csv("abuseipdb_resultados.csv")

# Filtrar solo las IPs maliciosas con AbuseScore >= 50 (ajusta si quieres usar otro umbral)
df_maliciosas = df_maliciosas[df_maliciosas["AbuseScore"] >= 100]

# Hacer el merge entre logs y IPs maliciosas usando src_ip e IP como claves
df_filtrado = df_logs.merge(df_maliciosas, left_on="Src. IP", right_on="IP", how="inner")

# Guardar el resultado si lo deseas
df_filtrado.to_csv("logs_con_ips_maliciosas.csv", index=False)

# Mostrar un resumen
print(f"Se encontraron {len(df_filtrado)} eventos con IPs maliciosas.")
df_filtrado.keys()
```

```python

# Cargar el archivo con IPs maliciosas
df_maliciosas = pd.read_csv("logs_con_ips_maliciosas.csv")

# Asegurarse de que ambas columnas de IP estén en formato string
df["src_ip"] = df["src_ip"].astype(str)
df_maliciosas["IP"] = df_maliciosas["IP"].astype(str)

# Añadir columna 'es_maliciosa': 0 si la IP está en la lista de maliciosas, 1 si no
df["es_maliciosa"] = df["src_ip"].apply(lambda ip: 0 if ip in df_maliciosas["IP"].values else 1)

# Verificar algunas filas como muestra
df[["src_ip", "es_maliciosa"]].head(10)
```
